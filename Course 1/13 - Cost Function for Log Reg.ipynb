```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from lab_utils_common import plot_data, sigmoid, dlc
plt.style.use('./deeplearning.mplstyle')
```

```python
# Simple dataset:
X_train = np.array([
    [0.5, 1.5],
    [1, 1],
    [1.5, 0.5],
    [3, 0.5],
    [2, 2],
    [1, 2.5]
])
y_train = np.array([0, 0, 0, 1, 1, 1])
# m = 6
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
plot_data(X_train, y_train, ax)
ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12, c='gray')
ax.set_xlabel('$x_2$', fontsize=12, c='gray')
plt.show()
```

```python
def compute_cost_logistic(X, y, w, b):
    # Vectorization, baby.
    m = len(y)
    cost = 0
    z = np.matmul(
        np.c_[X, np.ones(m)],
        np.r_[w, b]
    )
    f_wb = sigmoid(z)
    cost = (1 / m) * np.sum( -y * np.log(f_wb) - (1 - y) * np.log(1 - f_wb) )
    
    return cost
```

```python
w = np.array([1, 1])
b = -3
print(compute_cost_logistic(X_train, y_train, w, b))
# Expect ~0.367
```

```python
x0 = np.arange(0, 6)
x1 = 3 - x0
x1_other = 4 - x0

fix, ax = plt.subplots(1, 1, figsize=(4,4))
# Decision boundary:
ax.plot(x0, x1, c=dlc['dlblue'], label='$b$=-3')
ax.plot(x0, x1_other, c=dlc['dlmagenta'], label='$b$=-4')
ax.axis([0, 4, 0, 4])
plot_data(X_train, y_train, ax)
# ax.axis([0, 4, 0, 4])
ax.set_ylabel('$x_1$', fontsize=12, c='gray')
ax.set_xlabel('$x_0$', fontsize=12, c='gray')
plt.legend(loc='upper right')
plt.title("Decision Boundary")
plt.show()
```
