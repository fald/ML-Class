```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from lab_utils_common import dlc, plot_data
from plt_one_addpt_onclick import plt_one_addpt_onclick
plt.style.use('./deeplearning.mplstyle')
```

```python
# Sample data
x_train = np.array([x for x in range(6)])
y_train = np.array([0, 0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5],
                     [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos = y_train == 1
neg = y_train == 0

fig, ax = plt.subplots(1, 2, figsize=(8, 3))
ax[0].scatter(x_train[pos], y_train[pos], marker='x', c='r', s=80, label="y == 1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label="y == 0",
             facecolors="none", edgecolors=dlc["dlblue"], lw=3)
ax[0].set_ylim(-0.08, 1.1)
ax[0].set_ylabel('y', fontsize=12, c='gray')
ax[0].set_xlabel('x', fontsize=12, c='gray')
ax[0].set_title("One variable plot.")
ax[0].legend()

plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_ylabel("$x_1$", fontsize=12, c='gray')
ax[1].set_xlabel("$x_2$", fontsize=12, c='gray')
ax[1].set_title("Two variable plot.")
ax[1].legend()

plt.tight_layout()
plt.show()
```
