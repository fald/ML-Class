```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
z = np.arange(-10, 11)
y = sigmoid(z)

print(np.c_[z, y])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 3))
ax.plot(z, y, c='b')
ax.set_title("Sigmoid")
ax.set_ylabel("sigmoid(z)")
ax.set_xlabel("z")
draw_vthresh(ax, 0)
```
